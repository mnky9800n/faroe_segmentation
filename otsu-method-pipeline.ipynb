```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cv2
import helpers
```

```python
filename = 'benoitdata/F20_10_b/6um_jamtveit_att_pag-1.99_3.12_00445.tif'

img_f2010b = cv2.imread(filename,0)
```

```python
i_ves = helpers.apply_otsu_dilate_erode(img_f2010b)
i_ves = helpers.remove_convex_area_erode(i_ves)
i_ves = helpers.convert_to_binary(i_ves)

i_frac = helpers.label_fractures(img_f2010b)
i_frac = helpers.median_max(i_frac)
i_frac = helpers.remove_convex_area_eccentricity(i_frac)


fig, ax = helpers.plot_image(img_f2010b, cmap='Greys')
ax.imshow(i_ves, alpha=0.5, cmap='PRGn', interpolation='none')
ax.imshow(i_frac, alpha=0.75, cmap='plasma_r', interpolation='none')
```
